```python
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pickle
from keras import backend as K
from keras.optimizers import SGD, Adam

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import math

import joblib

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
    # Currently, memory growth needs to be the same across GPUs
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Memory growth must be set before GPUs have been initialized
        print(e)
```

```python
NSTOCK_DAYS = 30 #number of days of stock data, everything else assumed option data
NTARGET = 2 #number of targets to predict

#Read in the data made my PrepareOptionData
def loadXYData(ticker, corp='C', ns_back=20):
    #xydata = np.load('./optionchaindata/training/'+ticker+'_'+corp+'_XY_'+str(ns_back)+'.npy')
    #xydata = np.load('./optionchaindata/training/'+ticker+'_'+corp+'_XY.npy')
    #remove any rows with NaN
    #xydata = xydata[~np.isnan(xydata).any(axis=1)]
    xydata = np.load('./optionchaindata/training/'+ticker+'_'+corp+'_XY_'+str(ns_back)+'.npz')['xydata']
    return xydata 


def loadAndScaleData(ticker, corp='C', ns_back=20):
    xydata = loadXYData(ticker, corp)
    
    x_file, y_file = xydata[:,:-NTARGET], xydata[:,-NTARGET:]
    x_optiondata = x_file[:,:-NSTOCK_DAYS]
    x_stockdata = x_file[:,-NSTOCK_DAYS:]
    y_targetprices = y_file
    
    scaler_optiondata = MinMaxScaler(feature_range=(0,1))
    x_optiondata_sc = scaler_optiondata.fit_transform(x_optiondata)
    
    scaler_stockdata = MinMaxScaler(feature_range=(0,1))
    x_stockdata_sc = scaler_stockdata.fit_transform(x_stockdata)

    scaler_targetprices = MinMaxScaler(feature_range=(0,1))
    y_targetprices_sc = scaler_targetprices.fit_transform(y_targetprices)
    
    xydata_sc = np.concatenate([x_optiondata_sc,x_stockdata_sc,y_targetprices_sc], axis = 1)
    
    return xydata_sc, [scaler_optiondata,scaler_stockdata,scaler_targetprices]

def makeHashingLayer(tickerlist):

    hashinglayer = keras.layers.StringLookup(output_mode='one_hot')
    hashinglayer.adapt(tickerlist)
    print('String hashing layer trained')
    return hashinglayer

def convertContractToTicker(contracts):
    ser = pd.Series(contracts)
    ser.replace('\d(C|P)\d', '',regex=True, inplace=True) #remove any C or P with a digit on either side
    ser.replace('\d+', '',regex=True, inplace=True) #remove all digits (to keep ticker)
    return np.array(ser.values,dtype=str)


cad_tickers = ['CGC','CM','RY','TD','BNS','BMO','BAM','SHOP','CP','CNI','MO','ENB','MFC','GOLD','TU','SLF','IMO','RCI','SU','CNQ']

us_tickers = ['HSBC','GLD','AAPL','MSFT','TSLA','AMZN','GOOGL', 'NVDA','BAC', 'PG', 'HD', 'MA', 'PFE', 'MRK','INTC', 'UPS', 'NFLX', 'RTX','AMD','CVS','PYPL','UL','C','RIO','BA','GS','CAT','GSK','BP','SHEL','VALE']

all_tickers = us_tickers + cad_tickers

HashingLayer = makeHashingLayer(all_tickers)
NCNAME = len(HashingLayer.get_vocabulary())
```

```python
#Load the data 
#tickers = ['NFLX','MSFT','NVDA']
#tickers = ['HSBC','GLD','AAPL','MSFT','TSLA','AMZN','GOOGL', 'NVDA','BAC', 'PG', 'HD', 'MA', 
#           'PFE', 'MRK','INTC', 'UPS', 'NFLX', 'RTX','AMD','CVS','PYPL','UL','C','RIO','BA','GS','CAT','GSK']

#Remove a few tickers for independent testing (had them originally to train the vocabulary for the string
#encoding layer)
tickers = all_tickers
tickers.remove('NFLX')
tickers.remove('HSBC')
tickers.remove('NVDA')

xydata = np.array([])

for i in range(len(tickers)):
    ticker = tickers[i]
    if i == 0:
        xydata = loadXYData(ticker,'C', NSTOCK_DAYS) 
    else:
        data =  loadXYData(ticker,'C', NSTOCK_DAYS)
        xydata = np.concatenate([xydata,data])


#Perform a shuffle to remove time correlations and mix data sets
np.random.shuffle(xydata)
x_file, y_file = xydata[:,:-NTARGET], xydata[:,-NTARGET:]


#First NOPTION_DATA columns of x are:
#strike, openInterest, daysToExpiry, deltaStrike,impliedVolatility, deltaDays, weekday
#then the stock data. We want to apply different normalizations to them

ntest = int(0.1*x_file.shape[0])


#The contract names get hashed:
x_cname = convertContractToTicker(x_file[:,0]) #HashingLayer(x_file[:,0]) #first column is contract name
x_optiondata = x_file[:,1:-NSTOCK_DAYS]
x_stockdata = x_file[:,-NSTOCK_DAYS:]
y_targetprices = y_file

#
#Scale the data
scaler_stockdata = MinMaxScaler(feature_range=(0,1))
x_stockdata_sc = scaler_stockdata.fit_transform(x_stockdata)

scaler_optiondata = MinMaxScaler(feature_range=(0,1))
x_optiondata_sc = scaler_optiondata.fit_transform(x_optiondata)

scaler_targetpricesprices = MinMaxScaler(feature_range=(0,1))
y_targetprices_sc = scaler_targetpricesprices.fit_transform(y_targetprices)

#Reshape to add a dimension for feature
x_cname_sc = np.reshape(x_cname, (x_cname.shape[0],1))
x_optiondata_sc = np.reshape(x_optiondata_sc, (x_optiondata_sc.shape[0],x_optiondata_sc.shape[1],1))
x_stockdata_sc = np.reshape(x_stockdata_sc, (x_stockdata_sc.shape[0],x_stockdata_sc.shape[1],1))

#Split training and testing
x_cname_train = x_cname_sc[:-ntest]
x_opdata_train = x_optiondata_sc[:-ntest]
x_stdata_train = x_stockdata_sc[:-ntest]
y_train = y_targetprices_sc[:-ntest]

print("training:",x_cname_train.shape, x_opdata_train.shape,x_stdata_train.shape)

x_cname_test = x_cname_sc[-ntest:]
x_opdata_test = x_optiondata_sc[-ntest:]
x_stdata_test = x_stockdata_sc[-ntest:]
y_test = y_targetprices_sc[-ntest:]

print("testing:",x_cname_test.shape, x_opdata_test.shape,x_stdata_test.shape)

print(y_test)



```

```python

NOPTION_DATA = x_opdata_test.shape[1]

model_name = 'sys_notick_conv_sm_'+str(NSTOCK_DAYS)+'d'

print("saving scaling parameters")
joblib.dump(scaler_stockdata, 'models/'+model_name+"_scaler_stockdata.save") 
joblib.dump(scaler_optiondata, 'models/'+model_name+"_scaler_optiondata.save") 
joblib.dump(scaler_targetpricesprices, 'models/'+model_name+"_scaler_targetprices.save") 

def make_model(name=model_name):
    
    inputCName = keras.Input(shape=(1), dtype=tf.string)
    inputOptionData = keras.Input(shape=(NOPTION_DATA, 1))
    inputStockData = keras.Input(shape=(NSTOCK_DAYS, 1))
    
    # X0 (the contract name, a string)
    # Convert to one_hot encoded
    #x0 = layers.Flatten()(inputCName)
    x0 = HashingLayer(inputCName)
    x0 = layers.Flatten()(x0)
    x0 = layers.Dense(16, activation="relu")(x0)
    x0 = layers.Dense(4, activation="relu",name='embedding_output')(x0)
    
    # X1
    #Flatten the option data
    x1 = layers.Flatten()(inputOptionData)
    x1 = layers.Dense(32, activation="relu")(x1)
    x1 = layers.Dense(8, activation="relu",name='option_dense_output')(x1)
    #x1 = layers.Dropout(0.1)(x1)
        
    # X2
    #GRU on the stock price data
    x2 = layers.GRU(units = 32, return_sequences = True, activation='tanh')(inputStockData)
    x2 = layers.Dropout(0.2)(x2)
    
    x2 = layers.GRU(units = 32, return_sequences = True, activation='tanh')(x2)
    x2 = layers.Dropout(0.2)(x2)    

    x2 = layers.GRU(units = 32, return_sequences = True, activation='tanh')(x2)
    x2 = layers.Dropout(0.2)(x2) 
    
    x2 = layers.GRU(units = 32, return_sequences = True, activation='tanh')(x2)
    x2 = layers.Dropout(0.2)(x2) 
    
    x2 = layers.Flatten()(x2)
    x2 = layers.Dense(32, activation="relu")(x2)
    x2 = layers.Dense(8, activation="relu",name='stock_gru_output')(x2)
    #x2 = layers.Dropout(0.1)(x2)
    
    
    # X3
    #Convo layer on stock price data
    x3 = layers.Conv1D(32, kernel_size = 1, strides=1, activation="relu")(inputStockData)
    x3 = layers.Conv1D(32, kernel_size = 3, strides=1, activation="relu")(x3)
    x3 = layers.Conv1D(32, kernel_size = 5, strides=1, activation="relu")(x3)
    x3 = layers.MaxPooling1D(2)(x3)
    x3 = layers.Conv1D(32, kernel_size = 3, strides=1, activation="relu")(x3)
    x3 = layers.MaxPooling1D(2)(x3)
    x3 = layers.Flatten()(x3)
    x3 = layers.Dense(32, activation="relu")(x3)
    x3 = layers.Dense(8, activation="relu",name='stock_convo_output')(x3)
    #x3 = layers.Dropout(0.1)(x3)
    
    # X4
    # Dense layer on the stock data
    x4 = layers.Flatten()(inputStockData)
    x4 = layers.Dense(32, activation="relu")(x4)
    x4 = layers.Dense(8, activation="relu",name='stock_dense_output')(x4)
    
    #Combine the option and stock data
    x = layers.Concatenate(name='concat_output')([x1, x3, x4])

    x = layers.Dense(16, activation="relu")(x)
    #x = layers.Dropout(0.1)(x) 
    x = layers.Dense(NTARGET)(x)
    
    outputs = x
    
    model = keras.Model([inputCName,inputOptionData,inputStockData], outputs, name=name)
    #model = keras.Model(inputOptionData, outputs, name=name)
    model.summary()   
    
    return model






grunet = make_model()

#grunet.compile(optimizer=Adam(learning_rate=0.0001, #default 0.001
#                              beta_1=0.9,beta_2=0.999, epsilon=1e-07,amsgrad=False),
#              loss='mean_squared_error',
#              metrics=['loss','val_loss'])

grunet.compile(optimizer=Adam(learning_rate=0.0005),loss='mean_squared_error')
```

```python
print("training",model_name)

model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='models/'+model_name, save_weights_only=False,
    monitor='val_loss', mode='min', save_best_only=False,save_freq='epoch')

history = grunet.fit([x_cname_train,x_opdata_train, x_stdata_train], y_train, epochs=30, batch_size=64, validation_split=0.2,
                     shuffle=True,callbacks = [model_checkpoint_callback])

with open('models/hist_'+model_name, 'wb') as file_pi:
    pickle.dump(history.history, file_pi)

plt.plot(history.history['loss'],label = 'training loss')
plt.plot(history.history['val_loss'], label = 'validation loss')
plt.legend()
plt.show()
```

```python


#Run it on the test data
predictions = grunet.predict([x_cname_test,x_opdata_test, x_stdata_test])
print(predictions)
```

```python
def make_plots(x_opdata_test_rc, x_stdata_test_rc, pred, actual):
    diffs = pred - actual

    diffspct = diffs[:,0]/actual[:,0]

    n5pct = np.count_nonzero( (diffspct > -0.1) & (diffspct < 0.1) )


    print(n5pct/diffspct.shape[0],"percent of prices within 10%")


    plt.plot(actual[:,0],pred[:,0],'.',alpha=0.05)
    plt.xlabel("actual bid")
    plt.ylabel("predicted bid")
    plt.grid()
    plt.show()

    plt.hist(diffspct, bins=np.linspace(-0.5,0.5,100))
    plt.xlabel("pct error in bid")
    plt.grid()
    plt.show()

    #strike, stockPrice, stockVolume, openInterest, daysToExpiry, impliedVolatility, deltaDays, weekday
    plt.plot(x_opdata_test_rc[:,0],diffspct,'.',alpha=0.05)
    plt.xlabel("strike")
    plt.ylabel("pct error in bid")
    plt.ylim(-1,1)
    plt.grid()
    plt.show()
    
        #strike, stockPrice, stockVolume, openInterest, daysToExpiry, impliedVolatility, deltaDays, weekday
    plt.plot(x_opdata_test_rc[:,1],diffspct,'.',alpha=0.05)
    plt.xlabel("stock price")
    plt.ylabel("pct error in bid")
    plt.ylim(-1,1)
    plt.grid()
    plt.show()

    #strike, stockPrice, stockVolume, openInterest, daysToExpiry, impliedVolatility, deltaDays, weekday
    plt.plot(x_opdata_test_rc[:,1]-x_opdata_test_rc[:,0],diffspct,'.',alpha=0.05)
    plt.xlabel("deltaStrike")
    plt.ylabel("pct error in bid")
    plt.ylim(-1,1)
    plt.grid()
    plt.show()


    #strike, stockPrice, stockVolume, openInterest, daysToExpiry, impliedVolatility, deltaDays, weekday
    plt.plot(x_opdata_test_rc[:,4],diffspct,'.',alpha=0.05)
    plt.xlabel("days to expiry")
    plt.ylabel("pct error in bid")
    plt.ylim(-1,1)
    plt.grid()
    plt.show()
```
